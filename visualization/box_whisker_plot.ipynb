```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
## create data for the bar plot

# data sizes
m = 30 # rows
n =  6 # columns

# generate data
data = np.zeros((m,n))

for i in range(n):
    data[:,i] = 30*np.random.randn(m) * (2*i/(n-1)-1)**2 + (i+1)**2 # data[:, i] = ...: all the i'th column of the dataset. f.e : data[:,0] - > All the first column of the dataset
# np.random.randn(m): Generates an array of m random values from a standard normal distribution (mean 0, standard deviation 1).

print(data)
```

```python
# now for the boxplot

plt.boxplot(data)
plt.show()

plt.boxplot(data.T)
plt.show()

# the orange line inside the plot is median (quartile 2)
# bottom of plot is quartile 1
# top of plot is (quartile 3)

# top bottom lines = max and min
# stars are outliers = further away from distribution
```

```python
# now with seaborn
sns.boxplot(data=data,orient='v')
plt.show()
```

```python
# or as a pandas data frame
df = pd.DataFrame(data,columns=['zero','one','two','three','four','five'])
sns.boxplot(data=df,orient='h')
plt.show()
```
