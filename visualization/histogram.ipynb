
## SECTION: Visualizing data
### Histograms



```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
## create data for the histogram

# number of data points
n = 1000

# generate data - log-normal distribution
data = np.exp( np.random.randn(n)/2 ) #natural exp will never give negative value
plt.plot(data, 's')
```

```python
# show as a histogram

# number of histogram bins
k = 40

plt.hist(data,bins=k)
plt.show()
```

```python
# another option
y,x = np.histogram(data,bins=k)

# bin centers
xx = (x[1:]+x[:-1])/2

plt.plot(xx,y)
plt.show()
```
