
## SECTION: Visualizing data
###  Line plots


```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
## create data for the plot

# number of data points
n = 1000

# generate log-normal distribution
data1 = np.exp( np.random.randn(n)/2 )
data2 = np.exp( np.random.randn(n)/10 )
data3 = np.exp( np.random.randn(n)/2 + 1 )

print(data1)
# print(data2)
# print(data3)
```

```python
## plots of their histograms

# number of histogram bins
k = 20

plt.hist(data1,bins=k)
plt.hist(data2,bins=k)
plt.hist(data3,bins=k)

plt.show()
```

```python
# histogram discretization for the datasets
y1,x1 = np.histogram(data1,bins=k)
xx1 = (x1[0:-1] + x1[1:]) / 2
y1 = y1 / sum(y1) # convert to probability

y2,x2 = np.histogram(data2,bins=k)
xx2 = (x2[0:-1] + x2[1:]) / 2
y2 = y2 / sum(y2) # convert to probability

y3,x3 = np.histogram(data3,bins=k)
xx3 = (x3[0:-1] + x3[1:]) / 2
y3 = y3 / sum(y3) # convert to probability



# show the plots
plt.plot(xx1,y1,'s-',label='data1')
plt.plot(xx2,y2,'o-',label='data2')
plt.plot(xx3,y3,'^-',label='data3')

plt.legend()
plt.xlabel('Value')
plt.ylabel('Probability')
plt.show()
```

HIstogram Discretization:

np.histogram(data, bins=k): This computes the histogram of a dataset (data1, data2, or data3) using k bins. It returns the histogram values (y1, y2, y3) and the bin edges (x1, x2, x3).

Bin Centers Calculation:
(x1[0:-1] + x1[1:]) / 2: Calculates the center of each bin for the first dataset (data1). The same logic applies to the second and third datasets.

Probability Conversion:
y1 = y1 / sum(y1): Converts the histogram values to probabilities for the first dataset (data1). The same logic applies to the second and third datasets.

Plotting:
plt.plot(xx1, y1, 's-', label='data1'): Plots the probability distribution function (PDF) for the first dataset using square markers.
plt.plot(xx2, y2, 'o-', label='data2'): Plots the PDF for the second dataset using circular markers.
plt.plot(xx3, y3, '^-', label='data3'): Plots the PDF for the third dataset using triangular markers.

Legend and Labels:
plt.legend(): Adds a legend to the plot.
plt.xlabel('Value'): Sets the x-axis label.
plt.ylabel('Probability'): Sets the y-axis label.
