```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
## create data for the bar plot

# data sizes
m = 30 # rows
n =  6 # columns

# generate data
data = np.zeros((m,n)) # This creates a 2D NumPy array (data) with dimensions m rows and n columns, initialized with zeros.

for i in range(n):
    data[:,i] = 30*np.random.randn(m) * (2*i/(n-1)-1)**2 + (i+1)**2 # data[:,i] => i is column here and : before it means all the rows in that column
    # (i+1)**2 --> quadratic expension of the plots :)
```

```python
## show the bars!

fig,ax = plt.subplots(1,3,figsize=(8,2))

# 'naked' bars
ax[0].bar(range(n),np.mean(data,axis=0))
ax[0].set_title('Bar plot')

# just the error bars
ax[1].errorbar(range(n),np.mean(data,axis=0),np.std(data,axis=0,ddof=1),marker='s',linestyle='')
ax[1].set_title('Errorbar plot')

# both
ax[2].bar(range(n),np.mean(data,axis=0))
ax[2].errorbar(range(n),np.mean(data,axis=0),np.std(data,axis=0,ddof=1),marker='.',linestyle='',color='k')
ax[2].set_title('Error+bar plot')

plt.show()
```

```python
## manually specify x-axis coordinates

xcrossings = [ 1, 2, 4, 5, 6, 9 ]

plt.bar(xcrossings,np.mean(data,axis=0))
plt.errorbar(xcrossings,np.mean(data,axis=0),np.std(data,axis=0,ddof=1),marker='.',linestyle='',color='k') #std = standart deviation
plt.title('Bar+error plot')

plt.show()
```

```python
## note about bars from matrices

# data are groups (rows) X property (columns)
m = [ [2,5,4,3], [1,1,8,6] ] # create matrix m : 2 rows and 4 columns

# create Subplots : This creates a 2x2 grid of subplots and returns a figure (fig) and an array of axes (ax) representing the individual subplots.
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(8,8)) 


# Plot Matrix as Image (Grouping by Rows):  This subplot visualizes the matrix m with rows as groups and columns as properties using the imshow function. Each element in the matrix corresponds to a pixel in the image.
ax[0,0].imshow(m)


# Create Pandas DataFrame and Plot Bar Chart (Grouping by Rows):
# This creates a pandas DataFrame from the matrix m and then plots a bar chart where each row in the DataFrame corresponds to a group, 
# and each bar in a group represents a property.
df = pd.DataFrame(m,columns=['prop 0','prop 1','prop 2','prop 3'])
df.plot(ax=ax[1,0],kind='bar') # ROW
ax[1,0].set_title('Grouping by rows')


# This subplot visualizes the transpose of the matrix m with columns as groups and rows as properties.
# now other orientation (property X group)
ax[0,1].imshow(np.array(m).T) #Plot Transposed Matrix as Image (Grouping by Columns):
df.T.plot(ax=ax[1,1],kind='bar') #COL
ax[1,1].set_title('Grouping by columns')

plt.show()


# So, in the context of how these subplots are used, you can say that ax[1, 0] is associated with grouping by rows,
#  and ax[1, 1] is associated with grouping by columns.
```
